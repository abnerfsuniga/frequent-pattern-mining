```python
import pandas as pd
from apyori import apriori
```

```python
data = pd.read_csv("data.csv", encoding = 'unicode_escape')
data = data.dropna()
description_cat = data["Description"].astype('category').cat
description_dict = dict(enumerate(description_cat.categories))
data["DescriptionCat"] = description_cat.codes
data.head()
```

```python
transactions = list(data.groupby("InvoiceNo")["DescriptionCat"].apply(list).to_dict().values())
display(transactions[0])
```

```python
for relation in apriori(transactions, min_support = 0.01, min_confidence = 0.2, min_lift = 3, min_length = 2):
    for rule in relation.ordered_statistics:
        antecedent = list(map(lambda x: description_dict[x], rule.items_base))
        consequent = list(map(lambda x: description_dict[x], rule.items_add))
        print(antecedent, "=>", consequent)
```
